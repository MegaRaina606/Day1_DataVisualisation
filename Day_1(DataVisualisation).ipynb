```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title("Line Plot")
plt.xlabel("x axis")
plt.ylabel("Y axis")
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
#comput the x and y coordinates for points on a sine curve
x = np.arange(0, 4*np.pi,0.1)
y = np.sin(x)
plt.title("Sine waveform")

#Plot the points using matplotlib
plt.plot(x,y,color='g')
plt.show()
```

```python
plt.plot(x,y,'go--',linestyle='dashed',linewidth=1,markersize=5)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
#Assignment Question
x=np.arange(0,10)
y=x*x
plt.xlabel('X-axis',color='green')
plt.ylabel('y-axis',color='blue')

plt.plot(x,y)
```

```python
x=np.arange(1,10)
y=3*x+10
##plt plot
plt.plot(x,y,'ro--')
plt.plot(x,y,'r.',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X-axis',color='blue')
plt.ylabel('Y-axis',color='blue')
plt.title('2d-Diagram',color='blue')

```

```python
#creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```
